```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([-1, -0.5, 0, 0.5, 1.0])
```

```python
def tl(x,q):
    return max(q*x, (q-1)*x)
```

```python
def tl_seq(X,q):
    temp = []
    for x in X:
        QL = tl(x,q)
        temp.append(QL)
        
    return temp
```

```python
Q05 = tl_seq(X,0.05)
Q25 = tl_seq(X,0.25)
Q50 = tl_seq(X,0.50)
Q75 = tl_seq(X,0.75)
Q95 = tl_seq(X,0.95)
```

```python
fig = plt.figure(figsize=(5,4))
plt.plot(X,Q05, label='5%', c='k')
plt.plot(X,Q25, label='25%', c='b')
plt.plot(X,Q50, label='50%', c='r')
plt.plot(X,Q75, label='75%', c='g', linestyle='--')
plt.plot(X,Q95, label='95%', c='orange', linestyle='--')
plt.xlabel("Prediction error")
plt.xticks(np.arange(-1,1.0001, step=0.5))
plt.xlim([-1,1])
plt.ylabel("Loss")
plt.yticks(np.arange(0,1.0001, step=0.2))
plt.ylim([0,1])
plt.legend()
plt.grid()
plt.savefig('../visualization/tiltedLoss.pdf',bbox_inches='tight')
plt.show()
```
